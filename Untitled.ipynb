```python
def play(steps=False, resetBoard=False):
    global turn
    global b1
    global wKmoved
    global bKmoved
    global wRmoved
    global bRmoved
    wKmoved = False
    bKmoved = False
    wRmoved = {'l':False,'r':False}
    bRmoved = {'l':False,'r':False}
    if not steps:
        b1 = board("chess board avoidchecktest.csv")
    if resetBoard:
        b1 = board("chess board avoidchecktest.csv")
        turn = 'w'
    count = 0
    over = False
    if steps:
        step = 0
    moveMade = False
    while True: 
        ### look at team ###
        team = b1.myTeam(turn)
        
        for piece in team.keys():
            #print(piece)
            if 'K' in piece:
                #print('found king')
                check = b1.dangerDetector(team[piece][0], turn)
                #print(check)
                if check:
                    ### the king is currently in check and must move (or we must put the other king in check)
                    ### or the check must be otherwise blocked TODO
                    print(turn + ' king is in check')
                    piece, move = b1.pickRandomMove(turn, specific=piece)
                    moveMade = True
                    if piece is 'done':
                        ### checkmate!
                        print(str(piece[0]) + ' king is in checkmate')
                        if turn is 'b':
                            print('white wins')
                        if turn is 'w':
                            print('black wins')
                        over = True
                if not check:
                    break
        if over:
            break
                    
        ### pick a random piece and move ###
        if not moveMade:
            piece, move = b1.pickRandomMove(turn)
        
        if 'P' in piece:
            pawn = True
        else:
            pawn = False
        
        if piece == 'done':
            print(piece + ' after ' + str(count) + ' turns')
            break
        
        ### non castle moves ###
        if 'castle' not in move:
            y = board.moveDict[move]
            ### prepare the start and end positions ###
            square = team[piece][0]
            start = square
            if len(y) == 3:
                newSpace = b1.walk(square, y[0])
                newSpace = b1.walk(newSpace, y[1])
            else:
                newSpace = b1.walk(square, y)
            end = newSpace

            ### detects if 'takes' needs to be said instead of 'to' ###
            if b1.assessSquare(start, end, start) == 'foe':
                totake = ' takes '
                endpiece = str(b1.recon(end))
            else:
                totake = ' to '
                endpiece = ''
                
            ### this chunk keeps track of stuff necessary for castling ###
            if b1.recon(start) == 'wK':
                wKmoved = True
            if b1.recon(start) == 'bK':
                bKmoved = True
            if b1.recon(start) == 'wR':
                if start == ('a',1):
                    wRmoved['l'] = True
                if start == ('h',1):
                    wRmoved['r'] = True
            if b1.recon(start) == 'bR':
                if start == ('a',8):
                    bRmoved['l'] = True
                if start == ('h',8):
                    bRmoved['r'] = True
                    
            ### this moves a piece ###
            #print(recon(start))
            print(start)
            print(end)
            print(str(b1.recon(start)) + str(start[0]) + str(start[1]) + str(totake) + str(endpiece) + str(end[0]) + str(end[1]))

            b1.body.at[newSpace[1], newSpace[0]] = b1.body.at[square[1], square[0]]
            b1.body.at[square[1], square[0]] = '0'
            
            if pawn:
                if turn == 'w':
                    if newSpace[1] == 8:
                        b1.promote(turn, newSpace)
                if turn == 'b':
                    if newSpace[1] == 1:
                        b1.promote(turn, newSpace)
        
        ### castle ###
        
        if 'castle' in move:
            if 'L' in move:
                side = 'l'
            if 'R' in move:
                side = 'r'
            b1.castle(turn, side)
            if team is 'w':
                wRmoved[side] = True
                wKmoved = True
            if team is 'b':
                BRmoved[side] = True
                BKmoved = True
                
        ### keeps track of whose turn it is ###
        q = 0
        if turn == 'b':
            turn = 'w'
            q = 1
        if turn == 'w' and q == 0:
            turn = 'b'
        #######################################
        
        #print(board)
        count += 1
        if steps:
            step += 1
            if step == steps:
                return 'paused after ' +  str(step) + ' steps'
        #time.sleep(0.5)
        
```
